```
df = pd.read_csv ("Flunel Analyst.csv")
df.head ()
```

```
df['created_at'] = pd. to_datetime(df['created_at'],errors= 'coerce')
```

```
df['user_key'] = df['user_key']. fillna(-1). astype(str)
df.loc[df['user_key'].isna(),'user_key'] = 'guest_' + df.index.astype(str)
```

```
funnel_steps = ["home", "product", "cart", "purchase"]
```

```
funnel = {}

for step in funnel_steps:
    unique_users = df[df['event_type'] == step]['user_key'].nunique()
    funnel[step] = unique_users

funnel_df = pd.DataFrame(list(funnel.items()), columns=['step', 'unique_users'])
funnel_df
```

```
funnel_counts = (
    df.groupby('event_type')['user_key']
    .nunique()
    .reindex(funnel_steps)
    .reset_index()
    .rename(columns={'event_type': 'step', 'user_key': 'unique_users'})
)

funnel_counts
```

```
import plotly.express as px

fig = px.funnel(
    funnel_counts,
    x='unique_users',
    y='step',
    color='step'  # supaya warnanya beda-beda
)
fig.show()
```
