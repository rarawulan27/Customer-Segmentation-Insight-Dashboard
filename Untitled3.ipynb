```python
df = pd.read_csv('/content/RARA 1 REKAPAN PENJUALAN SPAREPART HCM 2025-2026 - PENJUALAN SPAREPART 2026.csv', skiprows=2, header=0)
```

```python
mask = (
    df['TGL KELUAR BARANG/PERBAIKAN'].isna()&
    df['TGL TRANSAKSI '].isna()&
    df['NO. INVOICE'].isna()
)
print("Baris yang akan di hapus:",mask.sum())
```

```python
df = df[~mask].reset_index(drop=True)
```
